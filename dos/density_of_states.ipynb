```python

def calculate_dos_3d_free_electron(energy, effective_mass):
    """
    Calculates the 3D density of states for a free electron gas.

    Args:
        energy (float or np.array): Energy in Joules.
        effective_mass (float): Effective mass of the electron in kg.

    Returns:
        float or np.array: Density of states in J^-1 m^-3.
    """
    hbar = 1.0545718e-34  # Reduced Planck constant in J*s
    term_constant = (2 * effective_mass) / (hbar**2)
    dos = (1 / (2 * np.pi**2)) * (term_constant**(3/2)) * (energy**(1/2))
    return dos

# Example usage:
energy_range = np.linspace(0.01 * 1.602e-19, 1 * 1.602e-19, 100) # Energies in Joules (e.g., 0.01 eV to 1 eV)
electron_mass = 9.1093837e-31 # Mass of electron in kg
dos_values = calculate_dos_3d_free_electron(energy_range, electron_mass)

fig, ax = plt.subplots()
ax.plot(energy_range,dos_values)
ax.set_yscale('log')
ax.set_ylim([0,np.max(dos_values)])
# You can then plot dos_values against energy_range
```
